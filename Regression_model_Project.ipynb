```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler , OneHotEncoder
```

```python
insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance
```

```python
ct = make_column_transformer(
    (MinMaxScaler() , ["age","bmi","children"]),
    (OneHotEncoder(), ["sex","smoker","region"])
)
```

```python
X = insurance.drop(["charges"],axis=1)
Y = insurance["charges"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
ct.fit(x_train)

x_train_normalized = ct.transform(x_train)
x_test_normalized = ct.transform(x_test)
```

```python
# create a model
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model.compile(
    loss=tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(),
    metrics=["mae"]
)

```

```python
class MyCallback(tf.keras.callbacks.Callback):
  def on_epoch_end (self, epoch, logs={}):
    if (logs.get("loss") < 3550):
        print ("\nLoss is low so stop training")
        self.model.stop_training = True
```

```python
callback = MyCallback()
model.fit(x_train_normalized,y_train,epochs=200,callbacks=[callback])
```

```python
model.evaluate(x_test_normalized,y_test)
```
